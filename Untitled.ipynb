```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
import random 
from sklearn.model_selection import train_test_split
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Conv2D,MaxPool2D,Flatten,Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
from sklearn.metrics import confusion_matrix,classification_report


```

```python
df_train=pd.read_csv('fashion-mnist_train.csv')
df_test=pd.read_csv('fashion-mnist_test.csv')
```

```python
training=np.array(df_train )
testing=np.array(df_test)
training
```

```python
plt.imshow(training[10010,1:].reshape(28,28))
```

```python
import random 
i=random.randint(1,60000)
plt.imshow(training[i,1:].reshape(28,28))
```

```python
X_train=training[:,1:]/255#we divided by 255 to get normalized value
y_train=training[:,0]
```

```python
X_test=testing[:,1:]/255
y_test=testing[:,0]
```

```python
from sklearn.model_selection import train_test_split# create validate datasetin order to prevent overfitting 

```

```python
X_train,X_validate,y_train,y_validate=train_test_split(X_train,y_train,test_size=0.2,random_state=12345)
```

```python
X_train.shape,X_validate.shape,X_test.shape
```

```python
X_train=X_train.reshape(X_train.shape[0],*(28,28,1))
X_test=X_test.reshape(X_test.shape[0],*(28,28,1))
X_validate=X_validate.reshape(X_validate.shape[0],*(28,28,1))
```

```python
X_train.shape,X_validate.shape,X_test.shape
```

```python
#model evaluation 
cnn_model=Sequential()
```

```python
cnn_model.add(Conv2D(32,2,2,activation='relu',input_shape=(28,28,1)))
```

```python
cnn_model.add(MaxPool2D(pool_size=(2,2)))
```

```python
cnn_model.add(Flatten())
```

```python
cnn_model.add(Dense(output_dim=32,activation='relu'))
cnn_model.add(Dense(output_dim=10,activation='sigmoid'))
```

```python
cnn_model.summary()
```

```python
cnn_model.compile(loss='sparse_categorical_crossentropy',optimizer=Adam(learning_rate=0.001),metrics=['accuracy'])
```

```python
epochs=50
```

```python
cnn_model.fit(X_train,y_train,batch_size=512,epochs=epochs,verbose=1,validation_data=(X_validate,y_validate))
```

```python
#model evaluation and predictiom
evaluation = cnn_model.evaluate(X_test, y_test)
print('Test Loss : {:.3f}'.format(evaluation[0]),' & Test Accuracy : {:.3f}'.format(evaluation[1]))
```

```python
Predicted_classes = cnn_model.predict_classes(X_test)
Predicted_classes
```

```python
L = 6
W = 6
fig, axes = plt.subplots(L,W, figsize=(18,18))
axes = axes.ravel()

for i in np.arange(0, L*W):
    axes[i].imshow(X_test[i].reshape(28,28))
    axes[i].set_title('Prediction Class = {:0.1f}\n True Class = {:.1f}'.format(Predicted_classes[i],y_test[i]))
    axes[i].axis('off')
    
plt.subplots_adjust(wspace=.5)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test,Predicted_classes)
plt.figure(figsize=(15,10))
sns.heatmap(cm,annot=True,cmap='rocket')
plt.show()
```

```python


from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test,Predicted_classes)
plt.figure(figsize=(15,10))
sns.heatmap(cm/cm.sum(),annot=True,cmap='rainbow')
plt.show()
```

```python
Total_Misclassification = 1-cm.diagonal().sum()/cm.sum()
print('Total Misclassification: ',Total_Misclassification)
```

```python
#we observe 9% images are missclassified 
cr = classification_report(y_test,Predicted_classes)
print('Classification Report : ','\n',cr)
```

```python
#model improvement can be done
cnn_modell=Sequential()
cnn_modell.add(Conv2D(64,3,3,activation='relu',input_shape=(28,28,1)))
cnn_modell.add(MaxPool2D(pool_size=(2,2)))
cnn_modell.add(Flatten())
```

```python
cnn_modell.add(Dense(output_dim=64, activation='relu'))
```

```python
cnn_modell.add(Dense(output_dim=10,activation='sigmoid'))
```

```python
cnn_modell.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
```

```python
epochs=50
```

```python
cnn_modell.fit(X_train,y_train,verbose=1,batch_size=512,epochs=epochs,validation_data=(X_validate,y_validate))
```

```python
evaluation1=cnn_modell.evaluate(X_test,y_test)
print('Test Loss : {:.3f}'.format(evaluation1[0]),' & Test Accuracy : {:.3f}'.format(evaluation1[1]))
Predicted_classes1=cnn_modell.predict_classes(X_test)
from sklearn.metrics import confusion_matrix,classification_report
cm1 = confusion_matrix(y_test,Predicted_classes1)
plt.figure(figsize=(15,10))
sns.heatmap(cm1/cm1.sum(),annot=True,cmap='rainbow')
plt.show()
# Sum the diagonal to get the total correctly predicted values
Total_Misclassification1 = 1-cm1.diagonal().sum()/cm1.sum()
print('Total Misclassification: ',Total_Misclassification1)
cr1 = classification_report(y_test,Predicted_classes1)
print('Classification Report : ','\n',cr1)
```
